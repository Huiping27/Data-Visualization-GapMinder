```python
# average population of a European country in 1976 compare to 2015

#filter the data in 1976 of European country

data_1976_europe = data[(data['year'] == 1976) & (data['continent'] == 'Europe')]
avg_pop_1976 = data_1976_europe['population'].mean()
avg_pop_1976
```

```python
#  What is the highest population a continent ever had?
# Hint: group by multiple columns

highest_pop_continent=data.groupby(['continent', 'population']).max()
highest_pop_continent
```

```python
# 9. Which continent had that population and in which year?
# Hint: group by multiple columns and filter for the result from #6

#Group the data by 'continent' and 'year' and find the maximum population for each continent

max_pop_continent=data.groupby(['continent', 'year', 'population']).max()
max_pop_continent
```

```python
# 10. Plot a bar plot comparison of life_expectancy, fertility, population average per continent
# Hint: for a quick and dirty graph use Pandas ploting ability
# Hint: use log=True to make the visual more friendly to the eyes

# Calculate the average life expectancy, fertility, and population per continent

continent_averages = data.groupby('continent').agg({'life expectancy': 'mean', 'fertility': 'mean', 'population': 'mean'})

continent_averages
```

```python
# reset the continent as index!

# firstly Reset index to make 'continent' a regular column
continent_averages.reset_index(inplace=True)

# Now set 'continent' as the index
continent_averages.set_index('continent', inplace=True)

```

# Part III

## Merge the average value to original DataFrame

```python
# Group data by continent and calculate the mean of each variable
continent_averages = data.groupby('continent').agg({
    'life expectancy': 'mean',
    'fertility': 'mean',
    'population': 'mean'
}).reset_index()  # Reset index to keep 'continent' as a column

# Merge continent_averages with the original DataFrame 'data'
data_merged = data.merge(continent_averages, on='continent', how='left')
```
